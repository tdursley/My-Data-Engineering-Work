# Test for Normality
Create a Q-Q Plot, Histogram or KDE for PM2.5 and AQI, then run a Shapiro Wilk test

```python
pm25_values = df["PM2.5"].dropna()

plt.figure(figsize=(6, 6))
stats.probplot(pm25_values, dist="norm", plot=plt)
plt.title("QQ Plot of PM2.5")
plt.show()

# Perform Shapiro-Wilk normality test
shapiro_test = stats.shapiro(pm25_values)
print(f"Shapiro-Wilk Test Statistic: {shapiro_test.statistic}, p-value: {shapiro_test.pvalue}")
```
